```python
import pandas as pd
import matplotlib.pyplot as plt
suicide_rate_df = pd.read_csv("Outputs/suicide_rate_df.csv")
```

```python
# GDP PPP PerPopulation MATRIX
# Read file
file1 = pd.read_csv("Resources/gdp_ppp_pop.csv", encoding='cp1252')
gdp_temp_df = file1.drop(file1.columns[1:34], axis=1).round(0)
gdp_temp_df = gdp_temp_df.drop(gdp_temp_df.columns[29], axis=1)

print(gdp_temp_df.shape)    #check shape and heading
gdp_temp_df.head(1)
```

```python
# DATA CLEANING : Checking if names in the key column of the two matrix are the same and list all the difference 
df1 = suicide_rate_df
df2 = gdp_temp_df
common1 = pd.merge(df1,df2,on=['Country Name'], how='outer')
#check for data that not in each other list for name spelling difference
outcast1_df = common1[~common1['Country Name'].isin(df1['Country Name'])].filter(['Country Name'], axis=1).reset_index(drop=True)
outcast2_df = common1[~common1['Country Name'].isin(df2['Country Name'])].filter(['Country Name'], axis=1).reset_index(drop=True)
outcast_df = pd.concat([outcast1_df,outcast2_df], ignore_index=True, axis=1)
outcast_df.head()
```

```python
#change the name to make the key column align
gdp_ed_df = gdp_temp_df.replace(['Brunei Darussalam','Cabo Verde', 'Congo, Rep.', 'Congo, Dem. Rep.','Egypt, Arab Rep.',
                            'Micronesia, Fed. Sts.','Iran, Islamic Rep.','Kyrgyz Republic','Lao PDR','North Macedonia',
                            'Korea, Dem. People’s Rep.','West Bank and Gaza','St. Lucia','St. Vincent and the Grenadines',
                            'Slovak Republic','Korea, Rep.','Eswatini','Syrian Arab Republic','Bahamas, The','Gambia, The',
                            'Venezuela, RB','Virgin Islands (U.S.)','Yemen, Rep.'],                             
                           ['Brunei', 'Cape Verde', 'Congo', 'Democratic Republic of the Congo','Egypt',
                            'Federated States of Micronesia','Iran','Kyrgyzstan','Laos','Macedonia','North Korea',
                            'Palestine','Saint Lucia','Saint Vincent and the Grenadines','Slovakia',
                            'South Korea','Swaziland','Syria','The Bahamas','The Gambia','Venezuela',
                            'Virgin Islands, U.S.','Yemen'])
```

```python
#double check if data is already clean as expected
common = pd.merge(df1,gdp_ed_df,on=['Country Name'], how='outer')
outcast_ed1_df = common[~common['Country Name'].isin(df1['Country Name'])].filter(['Country Name'], axis=1).reset_index(drop=True)
outcast_ed3_df = common[~common['Country Name'].isin(gdp_ed_df['Country Name'])].filter(['Country Name'], axis=1).reset_index(drop=True)
outcast_ed_df = pd.concat([outcast_ed1_df,outcast_ed3_df], ignore_index=True, axis=1)
outcast_ed_df
```
