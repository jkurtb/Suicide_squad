## Correlation Coefficient & Linear Regression

```python
x_values = Countries_HappinessIndex['Life Ladder']
y_values = Countries_HappinessIndex['Log GDP per capita']

(slope, intercept, rvalue, pvalue,stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = 'y = ' + str(round(slope,2)) + 'x + ' + str(round(intercept,2))

plt.scatter(x_values, y_values)
plt.plot(x_values, regress_values, 'r-')
plt.annotate(line_eq, (4,11), fontsize=14, color='red')
plt.xlabel('Happiness')
plt.ylabel('GDP per capita')
plt.show()
```

```python
def linregres_func(x_values, y_values, x_label, y_label, ann_horizontal, ann_vertical):
                                                                            
    (slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
    
    regress_values = x_values * slope + intercept
    eq_line = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
    
    plt.scatter(x_values, y_values, c='lightseagreen')
    plt.plot(x_values, regress_values, 'r-')
    plt.annotate(eq_line, (ann_horizontal, ann_vertical), fontsize=11, color='red')
    plt.title(f" {x_label} vs {x_label} \n")
    plt.xlabel(x_values)
    plt.ylabel(y_values)
    print(f"The r-squared is: {rvalue}")
#     plt.savefig(f'figures/{savefile}')
    plt.show()
```

```python
linregres_func(Countries_HappinessIndex["Life Ladder"], Countries_HappinessIndex["Log GDP per capita"], 'Happiness', 'GDP', 4, 11)
```

```python
linregres_func(Countries_HappinessIndex['Life Ladder'], Countries_HappinessIndex['Social support'], 
               'Happiness', 'Social support', 5, 5)
```

```python
plt.scatter(Countries_HappinessIndex['Life Ladder'],Countries_HappinessIndex['Log GDP per capita'],marker='o', c='maroon',
            edgecolor= 'lightskyblue', alpha=0.5, s=30)
plt.title(f"Figure : Happiness vs. GDP per capital")
plt.xlabel(f"Happiness Score")
plt.ylabel(f"GDP per capita")
plt.grid(True)
# plt.savefig('../output_data/Fig2_LatitudeVsHumidity.png')
plt.show()
```
