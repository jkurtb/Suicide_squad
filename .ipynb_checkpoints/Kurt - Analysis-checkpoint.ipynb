```python
# Dependencies
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

from numpy import nan
from numpy import isnan
from sklearn.impute import SimpleImputer
```

# World Happiness Report for the Year 2005-2019

Define Countries of Interest and Coverage Years for Analysis

```python
# G20 (except E.U.) plus random guest countries
countries = ['Argentina','Australia','Bangladesh','Brazil','Canada','China','Denmark', 
                    'Finland', 'France',  'Germany', 'India', 'Indonesia', 'Iran', 'Italy', 
                    'Japan', 'Malaysia', 'Mexico', 'Philippines','Russia', 'Saudi Arabia', 'Singapore', 
                    'South Africa', 'South Korea', 'Turkey', 'United Kingdom', 'United States']
# Report coverage Y2005-2019 
years = ['2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014',
         '2015', '2016', '2017', '2018', '2019']
```

### Independent Variable - Happiness Score per Country

```python
# Load/Read World Happiness Data 
HappinessIndex_summary = "Data sets/From Vivi/HappinessIndex.csv"
HappinessIndex = pd.read_csv(HappinessIndex_summary)
HappinessIndex.set_index('Country name', inplace=True)
HappinessIndex.head()
```

```python
# Load/Read World Happiness Data 
Happiness_2005_2019 = "Data sets/Raw Data/Happiness Score/Happiness_2005_2019.csv"
world_happiness = pd.read_csv(Happiness_2005_2019 )
world_happiness.set_index('Country name', inplace=True)
world_happiness.drop(labels=['2005', '2006'], axis=1, inplace=True)
world_happiness.head()
```

#### Sampling, Imputation, and Display Dataset without NaN Values

```python
# # Load/Read World Happiness Data 
# Sampling_2007_2019 = "Data sets/From Vivi/HappyCountries.csv"
# Sample_HappinessIndex = pd.read_csv(Sampling_2007_2019)
# # Sample_HappinessIndex.set_index('Country name', inplace=True)
# Sample_HappinessIndex.head(1)
```

```python
sampling_happiness = world_happiness[world_happiness.index.isin(countries)]
sampling_happiness
```

```python
# Imputation of NaN or missing values through SimpleImputer from sklearn
values = sampling_happiness.values
imputer = SimpleImputer(missing_values=nan, strategy='mean')
transform_sampling_happiness = imputer.fit_transform(values)
# transform_sampling_happiness
```

```python
happiness_score = np.array(transform_sampling_happiness)
transformed_happiness = pd.DataFrame(data=happiness_score, index=[countries], columns=[years]).round(decimals=2)
transformed_happiness.head(5)
```

### Sampling and Simple Imputation

### GDP per capita

```python
# Load/Read GDP per capita Data
GDP_vivi = 'Data sets/From Vivi/HI_Log GDPpCap.csv'
GDP = pd.read_csv(GDP_vivi)
GDP.set_index('Country name', inplace=True)
GDP.drop(labels=['2005', '2006'], axis=1, inplace=True)
# GDP.head(1)
```

```python
# Countries of Interest 
sampling_GDP = GDP[GDP.index.isin(countries)]
```

```python
# Imputation of NaN or missing values through SimpleImputer from sklearn
values = sampling_GDP.values
imputer = SimpleImputer(missing_values=nan, strategy='mean')
transform_GDP = imputer.fit_transform(values)
# Generate new dataset without NaN values 
GDP_score = np.array(transform_GDP)
clean_GDP = pd.DataFrame(data=GDP_score, index=[countries], columns=[years]).round(decimals=2)
```

```python
# Calculate the mean 
clean_GDP['mean'] = clean_GDP.mean(axis=1).round(decimals=2)
# Reformat df
mean_gdp = clean_GDP.drop(labels=['2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014',
                                 '2015', '2016', '2017', '2018', '2019'], axis=1)
GDP_meanscores = mean_gdp.rename(columns= {'level_0': 'Country', 'mean': 'Avg GDP Score'})
```
