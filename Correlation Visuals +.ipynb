```python
# Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
import seaborn as sns
```

```python
# Load/Read Happiness Index or Ladder Score
pd.set_option('display.max_columns', None)
TGE_Sheri_v4 = "Data sets/TGE_Sheri_v4.csv"
other_2017 = pd.read_csv(TGE_Sheri_v4)
other_2017.head(1)
```

```python
# Economic Health vs Happiness Index
otherFactors_correlation = other_2017.loc[:, ['Happiness Index', 'Brain drain index', 'Unemployment rate', 'Security threats index',
                                  'Fragile state index', 'Poverty percent of population', 'Unemployment rate',
                                  'Physician perpop', 'Substance Abuse', 'Depression & Anxiety', 'Public spending on education',
                                  'Suicide', 'Blue Index', 'Tourist arrivals', 'Int tourism revenue', 'Average Age', 
                                  'Literacy rate', 'Sleep (min)' ]]
```

```python
otherFactors_correlation = otherFactors_correlation.corr(method='pearson').round(decimals=2)
otherFactors_correlation
```

```python
%matplotlib notebook
sns.set(font_scale=.65)
sns.heatmap(otherFactors_correlation, 
            xticklabels=otherFactors_correlation.columns,
            yticklabels=otherFactors_correlation.columns,
            cmap="RdPu",
            annot=True,
            linewidth=0.5)
plt.title('Happiness and Other Factors \n Correlation Heat Map \n ', fontsize= 10)
plt.tight_layout()
plt.savefig('figures/Happiness and other Factors Correlation Heat Map')
plt.show()
```
